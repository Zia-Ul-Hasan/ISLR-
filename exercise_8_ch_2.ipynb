```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**PART A**

```python
College= pd.read_csv('/content/College.csv')

print(College.head(4))
print('/n')
print(College.describe)
```

**PART B**

```python
College.rename(columns={"Unnamed: 0": "College"},inplace=True)
College.set_index("College", inplace=True)
print(College)
```

PART C

```python
College.describe()
```

**PART D**

```python
features=['Apps','Enroll','Top10perc']
pd.plotting.scatter_matrix(College[features])
```

**PART E**

```python
sns.boxplot(x='Private' , y='Outstate', data=College)
```

PART F

```python
College['Elite'] = pd.cut(College['Top10perc'], [0,50,100], labels=['No', 'Yes'])
College['Elite'].value_counts()
sns.boxplot(x='Outstate', y='Elite', data=College)
```

PART G

```python
features = ['Apps', 'Accept', 'Outstate', 'Grad.Rate']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(features):
    College[col].plot.hist(bins=np.random.randint(5, 16), ax=axes[i])
    axes[i].set_title(f"Histogram of {col}")
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
corr_matrix = College.drop(['Private', 'Elite'], axis=1).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

The College dataset shows that schools with more applications generally have higher enrollments and larger student bodies. Colleges with stronger students (from the top 10% or 25%) tend to spend more, charge higher tuition, and have better graduation rates. Higher out-of-state tuition is linked to more spending and stronger outcomes, while books and personal expenses show little effect.
